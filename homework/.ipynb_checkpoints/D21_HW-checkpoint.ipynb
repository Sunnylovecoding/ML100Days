# 範例重點

如何使用亂數, 資料集來操作


```python
import matplotlib.pyplot as plt
import numpy as np

#畫出完整的 cos 圖形
x = np.arange(0, 3 * np.pi, 0.1)
y = np.cos(x)

plt.plot(x, y)
plt.show()
```

# 散點圖: Scatter Plots
    
    顏色由（X，Y）的角度給出。

    注意標記的大小，顏色和透明度。

```python
n = 1024
X = np.random.normal(0,1,n)
Y = np.random.normal(0,1,n)

T = np.arctan2(Y,X)
plt.scatter(X,Y, s=75, c=T, alpha=.5)

plt.xlim(-1.5,1.5)
plt.ylim(-1.5,1.5)

plt.show()
```

```python
np.random.seed(21)
n = 1024
X = np.random.normal(0,1,n)
Y = np.random.normal(0,1,n)
cValue = ['r','y','g','b','r','y','g','b']
plt.scatter(X, Y, c = cValue, marker = 's')

plt.title('scatter plot')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```
