### 題目 : 
將以下問卷資料的職業(Profession)欄位缺失值填入字串'others'，更進一步將字串做編碼。

此時用什麼方式做編碼比較適合?為什麼?

```python
import pandas as pd
q_df = pd.DataFrame([['male', 'teacher'], 
              ['male', 'engineer'],
              ['female', None],
              ['female', 'engineer']],columns=['Sex','Profession'])
q_df
```

```python
#缺失值填入字串'others'
q_df = q_df.fillna('other')
```

```python
#更進一步將字串做編碼。 此時用什麼方式做編碼比較適合?為什麼?
dummy = pd.get_dummies(q_df[['Profession']])
new_df = pd.concat([q_df['Sex'], dummy], axis = 1)
new_df
```

Ans: 使用虛擬變數(dummy variable)的方法比較合適，因為職業類別並無順序性。

### 1. 根據題目給的 DataFrame 完成下列操作：
- 計算每個不同種類 animal 的 age 的平均數
- 計算每個不同種類 animal 的 age 的平均數
- 將資料依照 Age 欄位由小到大排序，再依照 visits 欄位由大到小排序
- 將 priority 欄位中的 yes 和 no 字串，換成是布林值 的 True 和 False

```python
import numpy as np
import pandas as pd
data = {
    'animal': ['cat', 'cat', 'snake', 'dog', 'dog', 'cat', 'snake', 'cat', 'dog', 'dog'],
    'age': [2.5, 3, 0.5, np.nan, 5, 2, 4.5, np.nan, 7, 3],
    'visits': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
    'priority': ['yes', 'yes', 'no', 'yes', 'no', 'no', 'no', 'yes', 'no', 'no']
}

labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']
df = pd.DataFrame(data, index=labels)
df
```

```python
df.groupby(['animal']).mean()
```

```python
df.sort_values(by = 'age')
```

```python
df.sort_values(by = 'visits', ascending = False)
```

```python
# 利用 map 
#df['priority'] = df['priority'].map({'yes':True, 'no':False})
# 利用 replace
df.replace(['yes', 'no'],[True, False])
```

### 2. 一個包含兩個欄位的 DataFrame，將每個數字減去
- 該欄位的平均數
- 該筆資料平均數

```python
df = pd.DataFrame(np.random.random(size=(5, 3)))
df
```

```python
df1 = df - df.mean(axis = 0)
df1
```

```python
df2 = df.apply(lambda x: x-df.mean(axis = 1))
df2
```
